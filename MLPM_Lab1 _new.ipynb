```python
%pylab inline

# Signal generators
def sawtooth(x, period=0.2, amp=1.0, phase=0.):
    return (((x / period - phase - 0.5) % 1) - 0.5) * 2 * amp

def sine_wave(x, period=0.2, amp=1.0, phase=0.):
    return np.sin((x / period - phase) * 2 * np.pi) * amp

def square_wave(x, period=0.2, amp=1.0, phase=0.):
    return ((np.floor(2 * x / period - 2 * phase - 1) % 2 == 0).astype(float) - 0.5) * 2 * amp

def triangle_wave(x, period=0.2, amp=1.0, phase=0.):
    return (sawtooth(x, period, 1., phase) * square_wave(x, period, 1., phase) + 0.5) * 2 * amp

def random_nonsingular_matrix(d=2):
    """
    Generates a random nonsingular (invertible) matrix if shape d*d
    """
    epsilon = 0.1
    A = np.random.rand(d, d)
    while abs(np.linalg.det(A)) < epsilon:
        A = np.random.rand(d, d)
    return A

def plot_signals(X, title):
    """
    Plot the signals contained in the rows of X.
    """
    figure().suptitle(title)
    for i in range(X.shape[0]):
        ax = plt.subplot(X.shape[0], 1, i + 1)
        plot(X[i, :])
        ax.set_xticks([])
        ax.set_yticks([])
        
import scipy.io.wavfile
def save_wav(data, out_file, rate):
    scaled = np.int16(data / np.max(np.abs(data)) * 32767)
    scipy.io.wavfile.write(out_file, rate, scaled)
    
#1.1
def make_mixtures(S,A):
    #X=AS
    X=np.dot(A,S)
    return X

#1.2
def plot_histograms(X, title):
    figure().suptitle(title)
    for i in range(X.shape[0]):
        ax = plt.subplot(X.shape[0], 1, i + 1)
        plt.hist(X[i,:])
        ax.axes.get_xaxis().set_ticks([])
        ax.axes.get_yaxis().set_ticks([])
    plt.show()

#1.3
def phi(n,a):
    return {
        0: ( -tanh(a) ),
        1: ( tanh(a) - a ),
        2: ( -np.power(a, 3) ),
        3: ( [-6*x for x in a] / (np.power(a,2) + 5) ),
    }.get(n,0)

def p(n,a):
    return {
        0: ( 1/cosh(a) ),
        1: ( np.exp([-x*x / 2 for x in a]) * cosh(a) ),
        2: ( np.exp([-x**4 / 4 for x in a]) ),
        3: ( np.exp([-(3 * x**4 + 30 * x*x)/2 for x in a]) ),
    }.get(n,0)
    
#1.4    
def whiten(X):
   from numpy import dot, sqrt, diag
   from numpy.linalg import eigh
   covMat = np.cov(X)
   d,V = eigh(covMat)
   D = diag(1./sqrt(d))
   W = dot(dot(V,D),V.T)
   Xwhiten = dot(W,X)
   return Xwhiten  

#1.5
def scatterplot_signals(X, title):
    import matplotlib.gridspec as gridspec
    figure().suptitle(title)
    gs = gridspec.GridSpec(X.shape[0], X.shape[0])
    for i in range(X.shape[0]):
        for j in range(X.shape[0]):
            ax = plt.subplot(gs[i,j])
            if(i != j):
                scatter(X[i,:], X[j,:], marker = '.') 
            else:
                text(0.5,0.5,i)
            ax.set_xticks([])
            ax.set_yticks([])

def plot_covariance(X, title):
    sigmaX=np.cov(X)
    figure().suptitle(title)
    imshow(sigmaX, cmap='gray', interpolation='nearest')

#1.7
def ICA(Xm, activation_function, learning_rate, iterations, num_sources,W):
    X = np.zeros(Xm.shape)
    for i in range(Xm.shape[0]):
        maxrow=max(Xm[i,:])
        X[i,:]=Xm[i,:]/maxrow
        #print maxrow
        
    for itera in range(iterations):
        a=W.dot(X)
        
        Z = activation_function(a)
        
        Xprime = W.T.dot(a)
        
        deltaW = np.zeros(W.shape)
        deltaW = learning_rate * ((W + Z.dot(Xprime.T))/X.shape[1])
        
        if abs(np.mean(deltaW))< 0.00000000001:
            print itera
            print abs(np.mean(deltaW))
            return W
        
        W+=deltaW
    return W
```

## 1.2 Histogram

The measurements tend to look more gaussian, compared to the source distributions, as can be seen in our plots. This can be explained through the central limit theoreme, which basically states that the sum of a set of random variables from different source distributions tends to look more gaussian after aggregation.
This is a prerequisite for the ICA, since we try to construct a demixing matrix as such, that it will split the signal into different sources, where the "gaussianity" of the source distribution is minimized.
## 1.6 Covariance

A covariant algorithm is an algorithm which is independent of its underlying unit quantification.
E.g. metric or imperial units.
This can be archieved through defining a "natural" metric which depends on distance in parameter space, or through taking into account the curvature of the objective function.
This is done in the case of the covariance ICA.

## 1.9 Kurtosis

Definition of Kurtosis:
$$\frac{  E[(x-\mu)^4]  }{(E[(x-\mu)^2])^2}-3$$

Since all pdf's symmetrical in respect to 0, $$\mu = 0$$ for all activation functions.

thus:
$$\frac{ E[x^4] }{ (E[x^2])^2 } - 3$$
it is:
$$E[g(x)] = \int_{-\infty}^{+\infty} g(x) * f(x) dx$$

1) 
$$\phi(x) = \frac{1}{cosh(x)}$$
normalization constant:
$$\frac{1}{\int^{\infty}_{-\infty}\frac{1}{cosh(x)}}=\frac{1}{\pi}$$

thus:
Kurtosis:
$$ \frac{\int_{-\infty}^{\infty} x^4 \frac{1}{cosh(x)}  \frac{1}{\pi}  dx   }{  (\int_{-\infty}^{\infty} x^2 \frac{1}{cosh(x)} \frac{1}{\pi}dx )^2  }$$
$$ = \frac{  \frac{5\pi^4}{16}  }{ (\frac{\pi^2} { 4})^2 } - 3 = 2 $$

2)
$$\phi(x) = -x + tanh(x)$$
normalization constant:
$$ \frac{1} { \int^{\infty }_{-\infty} e^{(-x^2/2)} cosh(x) dx} = \frac{1}{\sqrt{2\pi e}}$$ 

thus,
Kurtosis:

$$ \frac{\int_{-\infty}^{\infty}          x^4  e^{(-\frac{x^2}{2})} cosh(x)  \frac{1}{\sqrt{2\pi e}}      dx}{(\int_{-\infty}^{\infty}      x^2      e^{(-\frac{x^2}{2})} cosh(x)    \frac{1}{\sqrt{2\pi e}}  dx)^2} - 3$$
$$=\frac{10}{4} - 3 = -0.5$$

3)
$$\phi(x) = e^{-\frac{x^4}{4}}$$
normalization constant:
$$ \frac{1} { \int^{\infty }_{-\infty} e^{-\frac{x^4}{4}} dx} = \frac{\Gamma(\frac{1}{4})}{\sqrt{2}}$$ 

thus,
Kurtosis:

$$ \frac{\int_{-\infty}^{\infty}          x^4  e^{-\frac{x^4}{4}}   \frac{\Gamma(\frac{1}{4})}{\sqrt{2}}      dx}{(\int_{-\infty}^{\infty}      x^2      e^{-\frac{x^4}{4}}    \frac{\Gamma(\frac{1}{4})}{\sqrt{2}}   dx)^2} - 3$$

$$\frac{1}{\frac{4\Gamma(\frac{3}{4})^2}{\Gamma(\frac{1}{4})}} = \frac{1}{8\pi^2} \approx -2.987335$$

4)
$$\phi(x) = (x^2 + 5)^{-3}$$

$$ \frac{1} { \int^{\infty }_{-\infty} (x^2 + 5)^{-3} dx} = \frac{200\sqrt{5}}{3\pi}$$ 

thus,
Kurtosis:

$$ \frac{\int_{-\infty}^{\infty}          x^4  (x^2 + 5)^{-3}  \frac{200\sqrt{5}}{3\pi}      dx}{(\int_{-\infty}^{\infty}      x^2      (x^2 + 5)^{-3} \frac{200\sqrt{5}}{3\pi}  dx)^2} - 3$$
