### Notation

$\mathbf{X}$ is the $M \times T$ data matrix, containing $M$ measurements at $T$ time steps.

$\mathbf{S}$ is the $S \times T$ source matrix, containing $S$ source signal values at $T$ time steps. We will assume $S = M$.

$\mathbf{A}$ is the mixing matrix. We have $\mathbf{X} = \mathbf{A S}$.

$\mathbf{W}$ is the matrix we aim to learn. It is the inverse of $\mathbf{A}$, up to indeterminacies (scaling and permutation of sources).

$\phi$ is an elementwise non-linearity or activation function, typically applied to elements of $\mathbf{W X}$.

In the following assignments, you can make use of the signal generators listed below.



```python
%pylab inline

# Signal generators
def sawtooth(x, period=0.2, amp=1.0, phase=0.):
    return (((x / period - phase - 0.5) % 1) - 0.5) * 2 * amp

def sine_wave(x, period=0.2, amp=1.0, phase=0.):
    return np.sin((x / period - phase) * 2 * np.pi) * amp

def square_wave(x, period=0.2, amp=1.0, phase=0.):
    return ((np.floor(2 * x / period - 2 * phase - 1) % 2 == 0).astype(float) - 0.5) * 2 * amp

def triangle_wave(x, period=0.2, amp=1.0, phase=0.):
    return (sawtooth(x, period, 1., phase) * square_wave(x, period, 1., phase) + 0.5) * 2 * amp

def random_nonsingular_matrix(d=2):
    """
    Generates a random nonsingular (invertible) matrix if shape d*d
    """
    epsilon = 0.1
    A = np.random.rand(d, d)
    while abs(np.linalg.det(A)) < epsilon:
        A = np.random.rand(d, d)
    return A

def plot_signals(X):
    """
    Plot the signals contained in the rows of X.
    """
    figure()
    for i in range(X.shape[0]):
        ax = plt.subplot(X.shape[0], 1, i + 1)
        plot(X[i, :])
        ax.set_xticks([])
        ax.set_yticks([])
```

We see that the  histograms of the measurements tend to look like gaussians. The reason is that the distribution of the measurements is affected by the mixture matrix which is randomly generated. This is important for the ICA algorithm which cannot get any information for the direction of the columns of the mixing matrix A from two Gaussian densities. ### 1.4 Whitening (15 points)
Some ICA algorithms can only learn from whitened data. Write a method `whiten(X)` that takes a $M \times T$ data matrix $\mathbf{X}$ (where $M$ is the dimensionality and $T$ the number of examples) and returns a whitened matrix. If you forgot what whitening is or how to compute it, various good sources are available online, such as http://courses.media.mit.edu/2010fall/mas622j/whiten.pdf

```python
def whiten(X): 
    Xcov = cov(X) 
    d,V = eigh(Xcov) 
    D = diag(1./sqrt(d+0)) 
    W = dot(dot(V,D),V.T) 
    Xwhiten = dot(W,X) 
    return Xwhiten

wX = whiten(X)
```

### 1.7 Independent Component Analysis (25 points)
Implement the covariant ICA algorithm as described in MacKay. Write a function `ICA(X, activation_function, learning_rate)`, that returns the demixing matrix $\mathbf{W}$. The input `activation_function` should accept a function such as `lambda a: -tanh(a)`. Update the gradient in batch mode, averaging the gradients over the whole dataset for each update. Try to make it efficient, i.e. use matrix operations instead of loops where possible (loops are slow in interpreted languages such as python and matlab, whereas matrix operations are internally computed using fast C code).

```python
############## Batch ICA Method ########################################################################
def ICA(X, activation_function, learning_rate = 1.0 , threshold = 0.01 , print_interval = 200000000000):
    
    #Algorithm Variables
    N = X.shape[1]                           # Size of Observation Set.
    I = X.shape[0]                           # Size of each observation vector x.
    W = random_nonsingular_matrix(I)         # Starting Weight matrix.
    average_dW = zeros((I,I))                # Natural Gradient average
    
    #Helper Variables
    iterations = 0;
    threshold_check = 100;
    
    while threshold_check > threshold:
        
        # While the average dW is more than a set threshold do:
        ################################################################################################
        a = dot(W,X)                              # Put x through a linear mapping
        z = activation_function(a)                # Put a through a nonlinear map:
        xp = dot(W.T,a)                           # Put a back through W
        average_dW = W + dot(z,xp.T)/ N           # Calculate the average gradient for the whole dataset
        W += learning_rate*average_dW             # Adjust the weight table W
        ################################################################################################
        
        # Recalculate threshold and print at a set interval.
        iterations +=1
        threshold_check = sum(np.absolute(average_dW))
        if iterations%print_interval == 0 : print iterations, threshold_check
            
    return W

############## Online ICA Method ########################################################################
def onlineICA(X, activation_function, learning_rate = 1.0):
    # Algorithm Variables
    N = X.shape[1]                           # Size of Observation Set.
    I = X.shape[0]                           # Size of each observation vector x.
    W = random_nonsingular_matrix(I)         # Starting Weight matrix.
    dW = zeros((I,I))                        # Natural Gradient average
    
    # Helper variables
    count = 1.0
    check = 1;
    
    for x in X.T:
        a = dot(W,x)
        z = activation_function(a)
        xp = dot(W.T,a)
        for i in range(W.shape[0]):
            for j in range(W.shape[0]):
                dW[i][j] = W[i][j] + xp[j]*z[i]
        W = W + learning_rate*dW
    return W
```

### 1.9 Excess Kurtosis (20 points)
The (excess) kurtosis is a measure of 'peakedness' of a distribution. It is defined as
$$
\verb+Kurt+[X] = \frac{\mu_4}{\sigma^4} - 3 = \frac{\operatorname{E}[(X-{\mu})^4]}{(\operatorname{E}[(X-{\mu})^2])^2} - 3
$$
Here, $\mu_4$ is known as the fourth moment about the mean, and $\sigma$ is the standard deviation.
The '-3' term is introduced so that a Gaussian random variable has 0 excess kurtosis.
We will now try to understand the performance of the various activation functions by considering the kurtosis of the corresponding priors, and comparing those to the empirical kurtosis of our data.

First, compute analytically the kurtosis of the four priors that you derived from the activation functions before. To do this, you will need the normalizing constant of the distribution, which you can either obtain analytically (good practice!), using computer algebra software (e.g. Sage) or by numerical integration (see scipy.integrate).

Now use the `scipy.stats.kurtosis` function, with the `fisher` option set to `True`, to compute the empirical kurtosis of the dummy signals and the real audio signals. Can you use this data to explain the performance of the various activation functions?

### Analytical Kurtosis computation
All the pdfs we are studying have $\mu = 0$ so the kurtosis can be calculated as   $\verb+Kurt+[X] = \frac{\operatorname{E}[X^4]}{(\operatorname{E}[X^2])^2} - 3$ . The definite integrals needed for calculating the expectations were calculated using Wolfram.

---

$$p(a) \propto \frac{1}{cosh(a)}$$
The normalisation constant as calculated by scipy.integrate is $C = \frac{1}{\pi}$
$$
\operatorname{E}[a^4] = \int\limits_{-\infty}^{\infty} p(a) a^4 da  = \frac{1}{\pi} \int\limits_{-\infty}^{\infty} \frac{1}{cosh(a)} a^4 da = \frac{1}{\pi}  \frac{5 \pi^5}{16}  =  \frac{5 \pi^4}{16}
$$
$$
\operatorname{E}[a^2] = \int\limits_{-\infty}^{\infty} p(a) a^2 da  = \frac{1}{\pi} \int\limits_{-\infty}^{\infty} \frac{1}{cosh(a)} a^2 da =  \frac{1}{\pi}  \frac{\pi^3}{4}  =  \frac{\pi^2}{4}
$$
$$
\verb+Kurt+[X] = \frac{\operatorname{E}[X^4]}{(\operatorname{E}[X^2])^2} - 3 = \frac{ \frac{5 \pi^4}{16}}{( \frac{\pi^2}{4})^2} - 3 = 2
$$

---

$$p(a) \propto e^\frac{-a^2}{2} cosh(a)$$ 
The normalisation constant as calculated by scipy.integrate is $C = \frac{1}{\sqrt{2\pi e}}$
$$
\operatorname{E}[a^4] = \int\limits_{-\infty}^{\infty} p(a) a^4 da  = \frac{1}{\sqrt{2\pi e}} \int\limits_{-\infty}^{\infty} e^\frac{-a^2}{2} cosh(a) a^4 da = \frac{1}{\sqrt{2\pi e}} 10 \sqrt{2\pi e} = 10
$$
$$
\operatorname{E}[a^2] = \int\limits_{-\infty}^{\infty} p(a) a^2 da  = \frac{1}{\sqrt{2\pi e}} \int\limits_{-\infty}^{\infty} e^\frac{-a^2}{2} cosh(a) a^2 da =  \frac{1}{\sqrt{2\pi e}}  2 \sqrt{2\pi e} =  2
$$
$$
\verb+Kurt+[X] = \frac{\operatorname{E}[X^4]}{(\operatorname{E}[X^2])^2} - 3 = \frac{10}{2^2} - 3 = -0.5
$$

---

$$p(a) \propto e^\frac{-a^4}{4}$$
The normalisation constant as calculated by scipy.integrate is $C = \frac{\sqrt{2}}{\Gamma(1/4)}$
$$
\operatorname{E}[a^4] = \int\limits_{-\infty}^{\infty} p(a) a^4 da  = \frac{\sqrt{2}}{\Gamma(1/4)} \int\limits_{-\infty}^{\infty} e^\frac{-a^4}{4} a^4 da = \frac{\sqrt{2}}{\Gamma(1/4)} \frac{\Gamma(1/4)}{\sqrt{2}} = 1
$$
$$
\operatorname{E}[a^2] = \int\limits_{-\infty}^{\infty} p(a) a^2 da  = \frac{\sqrt{2}}{\Gamma(1/4)} \int\limits_{-\infty}^{\infty} e^\frac{-a^4}{4} a^2 da = \frac{\sqrt{2}}{\Gamma(1/4)} \Gamma(3/4)\sqrt{2} \approx  0.66
$$
$$
\verb+Kurt+[X] = \frac{\operatorname{E}[X^4]}{(\operatorname{E}[X^2])^2} - 3 = \frac{1}{0.66^2} - 3 \approx  -0.7
$$


---

$$p(a) \propto \frac{1}{(a^2 +5)^3}$$
The normalisation constant as calculated by scipy.integrate is $C = \frac{200\sqrt{5}}{3\pi}$
$$
\operatorname{E}[a^4] = \int\limits_{-\infty}^{\infty} p(a) a^4 da  = \frac{200\sqrt{5}}{3\pi} \int\limits_{-\infty}^{\infty}  \frac{1}{(a^2 +5)^3} a^4 da = \frac{200\sqrt{5}}{3\pi} \frac{3 \pi}{8\sqrt{5}} = 25
$$
$$
\operatorname{E}[a^2] = \int\limits_{-\infty}^{\infty} p(a) a^2 da  = \frac{200\sqrt{5}}{3\pi} \int\limits_{-\infty}^{\infty}  \frac{1}{(a^2 +5)^3} a^2 da = \frac{200\sqrt{5}}{3\pi} \frac{\pi}{40\sqrt{5}} = \frac{5}{3}
$$
$$
\verb+Kurt+[X] = \frac{\operatorname{E}[X^4]}{(\operatorname{E}[X^2])^2} - 3 = \frac{25}{(\frac{5}{3})^2} - 3 = 6
$$
